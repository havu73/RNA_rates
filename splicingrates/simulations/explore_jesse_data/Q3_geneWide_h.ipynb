This analysis is about how close the gene-wide elongation rate is to the true gene-wide elongation rate. We need to be able to capture this statistics first and foremost before we expected to understand anything else

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
result_fn = '/gladstone/engelhardt/lab/hvu/RNA_rates/data_from_jesse/ha_reorganized_data/max_coverage.csv.gz'
df = pd.read_csv(result_fn, header = 0, index_col = None, sep = '\t')
```

```python
# histogram of max coverage
coverage = df['max_coverage']
plt.hist(coverage, bins = 100)
plt.xlabel('max coverage')
plt.ylabel('count')
plt.title('Histogram of max coverage')
```

```python
# histogram of the true gene-wide elongation rate
true_rate = df['true_avg_h']
plt.hist(true_rate, bins = 100)
plt.xlabel('true gene-wide elongation rate')
plt.ylabel('count')
plt.title('Histogram of true gene-wide elongation rate')
```

```python
# scatter plots of true vs. estimated gene-wide elongation rate
pred_h = df['pred_avg_h']
true_h = df['true_avg_h']
nonnan_idx = ~np.isnan(pred_h)
corr = np.corrcoef(true_h[nonnan_idx], pred_h[nonnan_idx])[0,1]
print('correlation = %.2f' % corr)
plt.scatter(true_h, pred_h, s = 1, alpha = 0.5)
plt.xlabel('true gene-wide elongation rate')
plt.ylabel('estimated gene-wide elongation rate')
plt.title('Scatter plot of true vs. estimated gene-wide elongation rate')
plt.text(4, 3.5, 'correlation = %.2f' % corr)
plt.show()
```
