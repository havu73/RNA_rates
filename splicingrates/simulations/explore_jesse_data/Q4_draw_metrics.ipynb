```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
def plot_mse_by_method(df, bin_value=1):
    # Filter the dataframe
    filtered_df = df[(df['bin_h'] == bin_value)]
    # Create the plot
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=filtered_df, x='method', y='MSE')
    # Customize the plot
    plt.title(f'MSE by Method (bin_h={bin_value})')
    plt.xticks(rotation=45)  # Rotate x-axis labels if they're long
    plt.xlabel('Method')
    plt.ylabel('MSE')
    # Adjust layout to prevent label cutoff
    plt.tight_layout()
    return plt

```

```python
metrics_fn = '/gladstone/engelhardt/lab/hvu/RNA_rates/data_from_jesse/ha_reorganized_data/metrics_trueH_predH.csv.gz'
df = pd.read_csv(metrics_fn, header = 0, index_col = None, sep = '\t')
plot_mse_by_method(df, bin_value=2)
```
